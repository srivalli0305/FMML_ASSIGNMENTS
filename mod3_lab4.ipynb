# **Section 1: Tweaking the parameters**

Let's stat the lab by writing knn code on a cancer detection dataset


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer(as_frame=True)
```

```python
cancer.target_names
```

```python
cancer.frame
```

```python
X = cancer.data
y = cancer.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 17)
```

## Section 1.1: Changing Value of K
Let's try to run the knn with different values of knn and compare the results

```python
neighbors = []
scores_list = []

for k in range(1, 25):
    neighbors.append(k)
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train,y_train)
    scores = knn.score(X_test,y_test)
    scores_list.append(scores)
```

```python
Accuracy = [x*100 for x in scores_list]
# plot Accuracy versus k
plt.figure(figsize = (10, 6))
plt.plot(neighbors, Accuracy)
plt.xlim([0, 25])
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy in %')
plt.show()
```

```python
# determining the best k
max_acc = max(Accuracy)
optimal_ks = [neighbors[i] for i, j in enumerate(Accuracy) if j == max_acc]
print(max_acc)
print(optimal_ks)
```

In this case, we see that the ideal number of neighbours can take any one of the values in the list [10,12,13,14,22,24].  
However, this only tells us the effectiveness of the KNN classifier only if we have the same train-test split considered above. It still does not tell us how well the algorithm would perform if the test set were different. The results would have been different if we had split the data differently.

### AUC-ROC Curve

The AUC - ROC curve is a performance measurement for classification problems at various threshold settings. ROC is a probability curve and AUC represents the degree or measure of separability.
It tells how much the model is capable of distinguishing between classes.  
The diagonal in the curve divides the ROC space. Points above the diagonal represent good classification results (better than random); points below the line represent bad results (worse than random).    
Higher the AUC, the better the model is at predicting 0 classes as 0 and 1 classes as 1.

The ROC curve is plotted with the TPR (True Positive Rate) on the y-axis, and the FPR (False Positive Rate) on the x-axis.

```python
## AUC-ROC
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import matplotlib.pyplot as plt

def plot_roc(y_test,predicted):
  y_scores = predicted
  fpr, tpr, threshold = roc_curve(y_test, y_scores)
  roc_auc = auc(fpr, tpr)
  plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
  plt.legend(loc = 'lower right')
  plt.plot([0, 1], [0, 1],'r--')
  plt.xlim([0, 1])
  plt.ylim([0, 1])
  plt.ylabel('True Positive Rate')
  plt.xlabel('False Positive Rate')
  plt.title('Receiver Operating Characteristic')
  plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train,y_train)
predictions = knn.predict(X_test)
plot_roc(y_test,predictions)
```

# **Section 2: Cross Validation**

As you saw in the above two cases with different train-test splits, there is a risk of overfitting on the test set because the parameters can be tweaked until the estimator performs optimally. This way, knowledge about the test set can â€œleakâ€ into the model and evaluation metrics no longer report on generalization performance.  

Cross Validation is an important technique in Machine Learning to analyze the performance of a model on **unseen** data.  

There are different kinds of cross-validation methods: exhaustive and non-exhaustive.   
Exhaustive cross-validation methods are cross-validation methods which learn and test on all possible ways to divide the original sample into a training and a validation set.
Non-exhaustive cross validation methods do not compute all ways of splitting the original sample.

One of the most commonly cross-validation methods is **K-Fold Cross validation**. The procedure for K-Fold Cross Validation is as follows:  
1. Split the dataset into k groups (typically without shuffling)
2. For each unique group:  
  a. Take the group as a hold out or test data set  
  b. Take the remaining groups as a training data set  
  c. Fit a model on the training set and evaluate it on the test set  
  d. Retain the evaluation score and discard the model  
3. Summarize the performance of the model using the mean of model evaluation scores over all the folds.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 17)
```

```python
from sklearn.model_selection import KFold
from statistics import mean
kf = KFold(n_splits=10)
cv_scores = []
neighbors = []
for k in range(1,25):
  knn = KNeighborsClassifier(n_neighbors = k)
  neighbors.append(k)
  acclist = []
  for train_index, test_index in kf.split(X_train, y_train):
    x_train_fold, x_test_fold = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]
    knn.fit(x_train_fold, y_train_fold)
    acclist.append(knn.score(x_test_fold, y_test_fold))
  cv_scores.append(mean(acclist))
```

```python
Accuracy = [x*100 for x in cv_scores]

# plot misclassification error versus k
plt.figure(figsize = (10, 6))
plt.plot(neighbors, Accuracy)
plt.xlabel('Number of neighbors')
plt.ylabel('Mean Cross-validation accuracy in %')
plt.show()
```

```python
# determining the best k
max_acc = max(Accuracy)
optimal_ks = [neighbors[i] for i, j in enumerate(Accuracy) if j == max_acc]
print(optimal_ks)
```

Note that the accuracy in the above case is not the accuracy on the test set. We plot the mean of the accuracies obtained during cross-validation, when we considered different splits of the training set.  

From the plot above, we can see that when the value of k is 10 or 12, our KNN model generalizes well to unseen data. Performing such cross validation gives us an idea about the parameters to choose when using a machine-learning algorithm.

# **Section 3: Standardization**


Scaling the data that we have available is a recommended pre-processing step when working with many machine learning algorithms. The scale and distribution of the data drawn from the domain may be different for each variable. Differences in the scales across input variables may increase the difficulty of the problem being modeled, and the model may suffer from poor performance during learning and sensitivity to input values resulting in higher generalization error.  

Standardization (also called z-score normalization) is one such scaling technique that transforms your data such that the resulting distribution has a mean of 0 and a standard deviation of 1.

$$ \text{New (scaled) value} = {X - \mu \over \sigma}$$


```python
X = cancer.data
y = cancer.target
```

### Questions to Think About and Answer
1. Instead of using cross-validation, we could also split our dataset into three sets- train, validation, and test set. The training proceeds on the training set, after which evaluation is done on the validation set, and when the experiment seems to be successful, final evaluation can be done on the test set.  
What do you think are the pros/cons of this method ?  
2. Is K-Fold cross validation an exhaustive or non-exhaustive cross validation method ? Justify your answer.

ANSWER : Splitting the dataset into three sets - training, validation, and test set - is a common practice in machine learning, and it has its own set of pros and cons.

Pros: Simplicity: The training-validation-test split is straightforward and easy to understand. It's a simple and intuitive way to assess the performance of a model.

Computationally Efficient: It can be computationally more efficient than K-Fold cross-validation, especially when dealing with large datasets. Training the model once on the training set and evaluating it on the validation set is faster than training and evaluating K times in cross-validation.

Useful for Small Datasets: In cases where the dataset is small, using a portion for testing and validation can be crucial. Cross-validation may be computationally expensive or infeasible with very limited data.

Cons: Dependence on Split: The performance metrics can be highly dependent on how the data is split. If the split is not representative of the overall dataset, it might lead to overfitting or underfitting.

Limited Assessment: The test set is only used once, and therefore its utility for assessing the model's generalization to unseen data is limited. It provides a one-time evaluation, and there's a risk of overfitting to the test set if adjustments are made based on its performance.

Reduced Sensitivity to Hyperparameters: With a single train-validation-test split, the model's sensitivity to hyperparameters might not be fully captured. Cross-validation provides a more robust estimate of model performance across different subsets of the data.

Wastage of Data: One-third of the data is reserved for testing, which means less data is available for training the model. In situations where data is scarce, this can be a significant drawback.

ANSWER : K-Fold cross-validation is a non-exhaustive cross-validation method. In K-Fold cross-validation, the dataset is divided into K folds, and the model is trained and evaluated K times, each time using a different fold as the validation set and the remaining data as the training set. This process provides an estimate of the model's performance by averaging the results over the K iterations.

Here's the justification for why K-Fold cross-validation is considered non-exhaustive:

Subset Representation: Each fold serves as both a training set and a validation set exactly once during the K iterations. However, each instance in the dataset is included in the validation set exactly once. Therefore, the model is not exhaustively tested on every possible combination of instances during the cross-validation process.

Randomness in Splitting: The process of splitting the data into folds often involves some level of randomness (e.g., shuffling the data before splitting). This randomness introduces variability in the subsets used for training and validation across different folds. As a result, the model is not exhaustively trained and validated on all possible subsets of the data.

Limited Exploration of Hyperparameters: K-Fold cross-validation typically focuses on exploring the performance of a model with different splits of the data. While it provides a robust estimate of performance, it may not exhaustively explore the entire hyperparameter space. Grid search or random search for hyperparameter tuning can be combined with K-Fold cross-validation to enhance the search, but it still may not be exhaustive.

In contrast, exhaustive cross-validation methods would involve training and validating the model on all possible combinations of instances, which is often impractical due to computational constraints. K-Fold cross-validation strikes a balance between providing a robust estimate of performance and being computationally feasible for a wide range of datasets and models.
